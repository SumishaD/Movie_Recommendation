```python
#To Extract only genres name from gernes column 
#Problem is it in list of string
#To extract only name of genres we use a method
# def convert(obj):
#     L = []
#     for i in obj:
#         L.append(i['name'])
#     return L
#convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
#Error (it should not be as str)
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L
```

```python
fmovies['genres'] = fmovies['genres'].apply(convert)
```

```python
fmovies.head()
```

```python
fmovies['keywords'] = fmovies['keywords'].apply(convert)
```

```python
fmovies['overview'] = fmovies['overview'].apply(lambda x:x.split())
```

```python
fmovies['genres']=fmovies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
fmovies['keywords']=fmovies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
# fmovies['overview'] = fmovies['overview'].apply(lambda x:x.split())
```

```python
fmovies['tags'] = fmovies['overview'] + fmovies['genres'] + fmovies['keywords']
```

```python
new_df = fmovies[['id', 'title', 'tags']]
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 5000, stop_words='english')
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem (text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
    
```

```python
#Example
ps.stem('actions')
```

```python
ps.stem('action')
```

```python
ps.stem('act')
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('The Dark Knight Rises')
```

```python
recommend('Iron Man')
```
